# Opencv Basics - I

```python
# Edge Detection and Image Gradients
import cv2

# Read Image
img = cv2.imread("child.jpg")

# Convert Image BGR to GrayScale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow("frame1", img)
cv2.imshow("frame2", gray)
print(gray.shape)

# cv2.Canny(image, threshold1, threshold2)
# Computing Edge Gradient with canny Thresholds
# The edge Gradients that are greater than threshold2 -> considered as edges
# The edge Gradients that are below than threshold1 -> considered not as an edges

edges = cv2.Canny(gray,60,150)
cv2.imshow("frame3", edges)

# Sobel() - Edge Detection
# Convert Image BGR to RGB
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height, width, _ = img1.shape
# Sebel Edges ksize = odd value
x_sobel = cv2.Sobel(img1, cv2.CV_64F, 0, 1, ksize = 5)
y_sobel = cv2.Sobel(img1, cv2.CV_64F, 1, 0, ksize = 5)

cv2.imshow("Sobel_X", x_sobel)
cv2.imshow("Sobel_X=Y", y_sobel)

# wait for any key`fc
cv2.waitKey(0)
# Close all frames
cv2.destroyAllWindows()
```

# Opencv Basic III 

```python
import cv2
import numpy as np

# Read Image
img = cv2.imread("child.jpg")

# Smoothening of Image using Kernel Matrix
kernel = np.ones((5,5), np.float32) / 25
print(kernel)

kernel_1 = np.array([[-1, -1, -1],
                     [-1, 9, -1],
                     [-1, -1, -1]])

# Filter 2D
# source, ddepth, kernel matrix
smoothed = cv2.filter2D(img, -1, kernel)

smoothed1 = cv2.filter2D(img, -1, kernel_1)

# Gaussian Blur
# Kernel shape must be odd positive (3,3),(5,5),(7,7)
# sigmaX = std deviation
g_blur = cv2.GaussianBlur(img, (9,9), 1)
cv2.imshow("Gaussian Blur", g_blur)

# Median Blur = Median of Pixels values under selected shape 
# Central Value from sleected 11 pixels values
m_blur = cv2.medianBlur(img, 9)
cv2.imshow("Median Blur", m_blur)

avg_blur = cv2.blur(img, (9,9))
cv2.imshow("Average Blur", avg_blur)

cv2.imshow("Original", img)
cv2.imshow("Blur1", smoothed)
cv2.imshow("Sharpen", smoothed1)

# wait for any key`
cv2.waitKey(0)
# Close all frames
cv2.destroyAllWindows()
```

# Opencv_Noisey

```python
import cv2
import random

def noise_black_white(image):
    # Get the Dimensions of Image
    row, col = image.shape
    
    # randomnly pick pixel values in Image to color those random pixels with black and white color
    number_of_pixels = random.randint(1500, 20000)
    for i in range(number_of_pixels):
        # pick random y-coordinate
        y_coordinate = random.randint(0, row - 1)
        # pick random x-coordinate
        x_coordinate = random.randint(0, col - 1)
        # color random pixels with white color
        image[y_coordinate][x_coordinate] = 255
    
    number_of_pixels = random.randint(1500, 20000)
    for i in range(number_of_pixels):
        # pick random y-coordinate
        y_coordinate = random.randint(0, row - 1)
        # pick random x-coordinate
        x_coordinate = random.randint(0, col - 1)
        # color random pixels with white color
        image[y_coordinate][x_coordinate] = 0
    return image

image = cv2.imread("parrot.jpg", cv2.IMREAD_GRAYSCALE)

cv2.imshow('Original Image',image)
cv2.imshow('Noisy Image',noise_black_white(image))

# Storing Image Permanently
cv2.imwrite("E:/OpenCV/noisy_image.png", noise_black_white(image))

# wait for any key
cv2.waitKey(0)
# Close all frames
cv2.destroyAllWindows()
```

# Opencv - Mouse Clicks 

```python
import cv2
import numpy as np

circles = np.zeros((4,2), np.int)
counter = 0

def mousePosition(event, x, y, flags, params):
    global counter
    # Event = Left Mouse Click
    if event == cv2.EVENT_LBUTTONDOWN:
        circles[counter] = x, y
        counter += 1
        print(circles)
        
img = cv2.imread("cards.jpg")

cv2.imshow('Original',img)
cv2.setMouseCallback('Original', mousePosition)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# OpenCV - ROI

```python
import cv2
import numpy as np

circles = np.zeros((2,2), np.int)
counter = 0

def mousePosition(event, x, y, flags, params):
    global counter
    # Event = Left Mouse Click
    if event == cv2.EVENT_LBUTTONDOWN:
        circles[counter] = x, y
        counter += 1
        print(circles)

img = cv2.imread('pug.jpg')

roi = img[66:385,442:815]

cv2.imshow('Original', img)
cv2.setMouseCallback('Original', mousePosition)
cv2.imshow('ROI Image', roi)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
